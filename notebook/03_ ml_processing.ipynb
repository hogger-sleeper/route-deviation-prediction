```python
import pandas as pd

route_dataset_df = pd.read_pickle(
    "../data/processed/route_dataset.pkl"
)

print(route_dataset_df.head())
```

```python
ml_df = route_dataset_df.copy()
```

```python
day_mapping = {
    "Monday": 0,
    "Tuesday": 1,
    "Wednesday": 2,
    "Thursday": 3,
    "Friday": 4,
    "Saturday": 5,
    "Sunday": 6
}

ml_df["Day of Week"] = ml_df["Day of Week"].map(day_mapping)
```

```python
label_mapping = {
    "ND": 0,
    "D": 1
}

ml_df["Label"] = ml_df["Label"].map(label_mapping)
```

```python
feature_columns = [
    "Driver ID",
    "Country",
    "Week ID",
    "Day of Week",
    "Planned Stop Count",
    "Planned Distance"
]
```

```python
X_reg = ml_df[feature_columns]

y_reg = ml_df["RQS"]
```

```python
X_cls = ml_df[feature_columns]

y_cls = ml_df["Label"]
```

```python
print(X_reg.head())

print("\nRegression Target:")
print(y_reg.head())

print("\nClassification Target:")
print(y_cls.head())
```

```python
from sklearn.model_selection import train_test_split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)

print(X_train_reg.shape)
print(X_test_reg.shape)

X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls,
    y_cls,
    test_size=0.2,
    random_state=42,
    stratify=y_cls
)

print(X_train_cls.shape)
print(X_test_cls.shape)

print(y_train_cls.value_counts(normalize=True))

print()

print(y_test_cls.value_counts(normalize=True))
```

```python
print(X_train_reg.shape)
print(X_test_reg.shape)

print()

print(y_train_cls.value_counts(normalize=True))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_reg_scaled = scaler.fit_transform(
    X_train_reg
)

X_test_reg_scaled = scaler.transform(
    X_test_reg
)

X_train_cls_scaled = scaler.fit_transform(
    X_train_cls
)

X_test_cls_scaled = scaler.transform(
    X_test_cls
)

print(X_train_reg_scaled.shape)
print(X_test_reg_scaled.shape)
```
