```python
import pandas as pd

classification_results = pd.DataFrame({

    "Model": [
        "Random Forest",
        "MLP",
        "LSTM",
        "Attention"
    ],

    "Accuracy": [
        0.7780,
        0.7148,
        0.7644,
        0.7265
    ]

})

classification_results
```

```python
regression_results = pd.DataFrame({

    "Model": [
        "Random Forest",
        "LSTM",
        "Attention",
        "CNN"
    ],

    "MAE": [
        0.1044,
        0.1054,
        None,
        0.0985
    ],

    "RMSE": [
        0.1442,
        0.1558,
        None,
        0.1451
    ],

    "R2": [
        0.4225,
        0.3558,
        None,
        0.4153
    ]

})

regression_results
```

```python
regression_results = pd.DataFrame({

    "Model":[
        "Random Forest",
        "MLP",
        "LSTM",
        "Attention"
    ],

    "MAE":[
        0.1044,
        None,
        None,
        None
    ],

    "RMSE":[
        0.1442,
        None,
        None,
        None
    ],

    "R2":[
        0.4225,
        None,
        None,
        None
    ]

})

regression_results
```

```python
best_cls = classification_results.sort_values(
    by="Accuracy",
    ascending=False
)

print(best_cls)
```

```python
best_reg = regression_results.sort_values(
    by="R2",
    ascending=False
)

print(best_reg)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    classification_results["Model"],
    classification_results["Accuracy"]
)

plt.title(
    "Classification Model Comparison"
)

plt.ylabel(
    "Accuracy"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    regression_results["Model"],
    regression_results["R2"]
)

plt.title(
    "Regression Model Comparison"
)

plt.ylabel(
    "R² Score"
)

plt.show()
```
